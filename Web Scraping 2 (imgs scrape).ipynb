```python
import selenium
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
drive = webdriver.Firefox()
drive.get('https://www.flickr.com/search/?text=cats')
drive.maximize_window()
time.sleep(10)

#Define an initial value
temp_height=0
    
while True:
    #Looping down the scroll bar
    drive.execute_script("window.scrollBy(0,1000)")
    #sleep and let the scroll bar react
    time.sleep(10)
    #Get the distance of the current scroll bar from the top
    check_height = drive.execute_script("return document.documentElement.scrollTop || window.pageYOffset || document.body.scrollTop;")
    #If the two are equal to the end
    if check_height==temp_height:
        break
    temp_height=check_height
```

```python
soup = BeautifulSoup(drive.page_source,'html.parser')
all_div_tags = soup.find_all('div','view photo-list-photo-view awake')
print(len(all_div_tags))
```

```python
card = all_div_tags[0]
print(card)
```

```python
img = card.find('img')
print(img['src'])
```

```python
img_src = []
for alldiv in all_div_tags:
    imgs = alldiv.find('img')
    imgs = 'https:'+imgs['src']
    img_src.append(imgs)
    
```

```python
for allimgs in img_src:
    print(allimgs)
```

```python
import urllib.request
pic = 1

for i in img_src:
    try:
        print(i)
        urllib.request.urlretrieve(i,'C:\\picss\\cat'+str(pic)+".jpg")
        pic += 1
    except Exception as e:
        print(str(e))
```
